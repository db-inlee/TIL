# 04 - Graphical Causal Models


그래프 모델은 인과성의 언어이다.

시작점으로 잠재적 결과의 조건부 독립을 가정해보자. 이것은 인과추론을 할 때 참으로 만들어야하는 주요한 가정 중 하나이다.

$ (Y_0, Y_1) \perp T | X $

조건부 독립은 결과에 있어서 다른 변수들이 아닌 ***처치*** 에 의한 효과만을 측정할 수 있게 만들어준다.


ex)고전적 예시 - 아픈환자에 대한 약의 효과
- 심각하게 아픈 환자들만 약을 먹는다면, 약을 처방하는 것이 환자의 건강을 악화시키는 것으로 볼 수 있다 -> 병의 심각성이 약의 효과와 섞여버리기 때문
- 환자를 심각한 케이스와 덜 심각한 케이스로 나누고 각 하위그룹에서의 약의 효과를 분석한다면, 실제 효과에 대한 명확한 그림을 얻을 수 있다.

여기서 모집단을 feature에 따라 나누는 것을 우리는 'X에 대한 통제' 혹은 'X에 대한 조건화'라고 한다.

- 심각한 케이스에 대해 조건화를 함으로써, 처치 매커니즘이 무작위에 가까워진다.
- 심각한 케이스에 속한 환자들은 이 차원에 대해 똑같기 때문에 이상 병증의 심각성에 따르는것이 아니라 오직 우연에 의해 약을 받거나 받지 않게 된다.
- 처치가 그룹 내에서 무작위로 할당되면 처치는 잠재적 결과와 조건부 독립이 된다.

***독립과 조건부 독립은 인과 추론의 핵심***

인과 그래프 모형을 통해 무엇이 무엇을 유발하는지, 인과성이 작동하는 방식을 표현해보자

```python
import pandas as pd
import numpy as np
import graphviz as gr
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
g = gr.Digraph()
g.edge("Z", "X")
g.edge("U", "X")
g.edge("U", "Y")

g.edge("medicine", "survived")
g.edge("severeness", "survived")
g.edge("severeness", "medicine")

g
```

각 노드는 랜덤 변수이다. 우리는 화살표, 또는 엣지를 사용해서 하나의 변수가 다른 것의 원인임을 나타낸다. 

- 첫번쨰 그래프 Z가 X를 유발, U가 X, Y를 유발 
- 두번쨰 그래프 병의 심각성(severeness)는 약(medicine)과 생존(survived) 모두의 원인이 되고, 약(medicine)은 생존(survived)의 원인이 된다.


## 그래프 모델의 이해

아래와 같이 단순한 그래프를 살펴보자.

A가 B의 원인이고, B가 C의 원인이 된다. 혹은 Z의 원인인 Y의 원인이 X이다.

```python
g = gr.Digraph()
g.edge("A", "B")
g.edge("B", "C")

g.edge("X", "Y")
g.edge("Y", "Z")
g.node("Y", "Y", color="red")

g.edge("causal knowledge", "solve problems")
g.edge("solve problems", "job promotion")

g
```

의존성이 화살표의 방향을 따라 흐르는 걸 볼 수 있다. 직관적이진 않지만, 의존성이 대칭적인 것을 봐보자.

이해하기 쉬운 예제로, 우리가 비즈니스 문제를 풀기 위한 유일한 방법이 인과추론이고, 이 문제를 푸는게 승진하는 유일한 길이라고 가정해보자.

이는 즉, 인과지식(causal knowledge)이 승진(job promotion)에 영향을 미치는 문제를 푸는 것과 같다. 인과적 전문성이 클수록 당신이 승진할 가능성이 더 커진다. 또한, 승진 가능성이 클수록 인과 지식을 가지고 있을 가능성도 크다. 그렇지 않으면 승진하기 어렵다.

매개 변수에 대해 조건화를 해보자.

이 경우, 의존성이 차단된다.

즉, X와 Z는 Y가 주어졌을 때 서로 독립이다. 위의 2번째 그래프를 봐보자.

그래프에서 빨간색 Y는 조건화된 변수임을 가리킨다. 

다시 승진 문제로 돌아와보자면, 같은 이유로 회사원 아무개씨가 Solve Problem에 능숙하다는 것을 안다면, 아무개씨가 인과 추론에 대해 안다는 것(causal knolwedge)가 있다는 것은 더 이상 아무개의 승진(job promotion) 가능성에 대한 추가적인 정보를 주지 않는다.

수학적 표시 : $E[Promotion|Solve Problems, Causal knowledge] = E[Promotion|Solve Problems]$ 라고 할 수 있다.

이것의 역도 참이다. 

즉, 내가 아무개씨가 문제를 얼마나 잘 푸는지 안다면, 아무개씨의 승진 상태에 대해 아는 것은 아무개씨가 인과 추론에 대해 알 가능성에 대한 추가적인 정보를 제공하지 않는다.

일반적인 규칙으로, 매개변수 B를 조건화하면서 A에서 C로 향하는 직선에 흐르는 의존성이 차단된다.

이는 $A \not \perp C $ 그리고 $ A \perp C|B $ 로 표현할 수 있다. 




***------------------------------------------------------------------------------------------------------------------***

아래와 같은 fork 구조에 대해 생각해보자.

그래프에서 같은 변수가 두 개의 다른 변수의 원인이 된다.

이 경우, 의존성은 화살표의 역방향으로 흐르고, 우리는 ***backdoor path***를 가지게 된다. 

우리는 공통의 원인에 대해 조건화함으로써 backdoor path를 막고 의존성을 제거할 수 있다

```python
g = gr.Digraph()
g.edge("C", "A")
g.edge("C", "B")

g.edge("X", "Y")
g.edge("X", "Z")
g.node("X", color="red")

g.edge("statistics", "causal inference")
g.edge("statistics", "machine learning")

g
```

위 3번째 그래프처럼 statistics(통계학)이 causal inference(인과추론)과 machine learning(기계학습)을 더 잘 알게 하는 원인이라고 가정해보자.

만약 아무개씨의 통계학 지식 수준에 대해 알지 못하지만, 아무개씨가 인과추론에 능하다는 것을 안다면, 아무개씨가 기계학습에도 능할 가능성도 높다고 여길 것이다.

이는 아무개씨의 통계 지식수준을 모르더라도 아무개씨의 인과추론 지식을 통해 그것을 추론할 수 있기 때문이다. 만약 아무개씨가 인과추론을 잘 안다면 아마도 아무개씨는 통계학도 잘 알 것이고, 그럼 기계학습도 잘 알 가능성도 높을 것이다.

이제 만약, 통계학에 관한 지식을 조건화하면, 기계학습에 대한 지식 수준과 인과추론에 대한 지식 수준은 독립이 된다.

일반적인 규칙으로서, 하나의 공통된 원인을 갖는 두 변수는 의존이지만, 공통된 원인을 조건화하면 독립이 된다.

이는 $A \not \perp B$ 그리고 $ A \perp B|C $ 로 표현 가능하다.

***-------------------------------------------------------------------------------------------***



이제 아래와 같은 collider 구조를 살펴보자.

collider는 두 개의 화살표가 하나의 변수에서 충돌하는 경우이다.  이 경우, 우리는 두 변수가 같은 효과를 공유한다고 말할 수 있다

```python
g = gr.Digraph()
g.edge("B", "C")
g.edge("A", "C")

g.edge("Y", "X")
g.edge("Z", "X")
g.node("X", color="red")

g.edge("statistics", "job promotion")
g.edge("flatter", "job promotion")

g
```

위의 세번째 그래프를 살펴보자.

승진(job promotion)을 하기 위해 두 개의 길이 있다고 하자. 
하나는 통계학(statistics)을 잘 하는 것, 또 하나는 상사에게 아첨(flatter)을 잘 하는 것이다.

만약, 승진에 대해 조건화하지 않는다면, 즉, 당신이 승진할지 아닐지 전혀 알 수 없다면, 당신의 통계학 수준과 아첨하는 것은 서로 독립이 된다.

다시 말해, 통계에 얼마나 능한지는 아첨을 얼마나 잘하는지 알려주는 바가 없다.

다른 한편으로, 만약 승진을 했다면? 통계학 수준과 아첨수준에 대해서도 말해줄 수 있다.

만약 통계학 수준이 높지 않은 경우에는 아첨을 잘하는 사람일 것이다, 그렇지 않으면 승진을 못했을 것이기 때문.
반대로 아첨을 잘 못한다면, 통계에 능할 것이다.

이런 현상을 ***잘 설명한다(explaning away)*** 라고 부르는데, 그 이유는 하나의 원인이 이미 효과에 대해 설명함으로써 다른 원인의 가능성을 줄여주기 때문이다.

일반적인 규칙으로 collider에 대해 조건화하는 것은 의존 경로(dependence path)를 열어준다.

반대로 조건화하지 않는 것은 닫힌 채로 두는 것이다.

이는 $A \perp B$ 그리고 $A \not \perp B | C$ 로 표현할 수 있다.

***---------------------------------------------------------------------------------------***

이 세구조(1.기본, 2.fork, 3.collider)에 대해 앎으로써, 더 일반적이 규칙을 도출할 수 있다. 

다음과 같은 경우에만 경로가 차단된다고 할 수 있다.
- 경로에 조건화된 non colider가 포함된다
- 경로에 아직 조건화되지 않은 collider가 포함되고 조건화된 후손(descendants)을 가지지 않는다.

어떻게 그래프에서 의존성이 흐르는지에 대한 자료가 있다. 
이는 Mark Paskin의 [Stanford 발표](http://ai.stanford.edu/~paskin/gm-short-course/lec2.pdf)출처이다. 

끝부분에 선이 그어진 화살표는 독립을 의미하고, 끝에 선이 없는 화살표는 의존을 의미한다. 

![img](data/img/graph-flow.png)



```python
g = gr.Digraph()
g.edge("C", "A")
g.edge("C", "B")
g.edge("D", "A")
g.edge("B", "E")
g.edge("F", "E")
g.edge("A", "G")

g
```

마지막 예시로서, 다음의 인과 그래프에서 독립성과 의존성 관계를 찾아보자.

1. Is $D \perp C $?
2. Is $D \perp C|A $?
3. Is $D \perp C|G $?
4. Is $A \perp F $?
5. Is $A \perp F|E $?
6. Is $A \perp F|E, C $?

정답 : 
- 1. True : 조건화되지 않은 collider를 가지고 있기 때문
- 2. False : 조건화된 collider를 가지고 있기 때문
- 3. False : 조건화된 collider의 후손을 포함하고 있기 때문, G를 A에 대한 일종의 proxy로 볼 수 있다.
- 4. True : 조건화되지 않은 collider인 B->E<-F를 포함하고 있기 때문
- 5. False : 조건화된 collider는 B->E<-F를 포함하기 때문
- 6. True : 조건화된 collider는 B->E<-F 를 포함하지만, 조건화된 non colider도 포함하기 때문. E에 대해 조건화 하는 것은 경로를 열어주지만, C에 대한 조건화하는 것은 이를 다시 닫아버림.

인과 그래프 모형에 대해 아는 것은 인과 추론에서 발생하는 문제를 이해하게 해준다.

우리가 이전 장에서 보았듯 문제는 항상 편향(bias)로 귀결된다.

$
E[Y|T=1] - E[Y|T=0] = \underbrace{E[Y_1 - Y_0|T=1]}_{ATET} + \underbrace{\{ E[Y_0|T=1] - E[Y_0|T=0] \}}_{BIAS}
$

그래프 모형은 우리가 어떤 편향을 다루고 있는지, 그것을 바로잡으려면 어떤 도구가 필요한지 알 수 있게 해준다.

# 교란 편향

편향의 첫번째 중요한 원인은 ***교란변수*** 이다.

이것은 처치와 결과가 같은 원인을 공유할 때 발생한다.

예를 들어, 처치가 '교육'이고 결과가 '소득'이라고 가정해보자.

교육이 소득에 미치는 인과적 효과를 추정하기 어려운 이유는 두 가지 공통의 원인인 '지능'을 가지고 있기 때문이다.

인과 효과를 규명하려면, 처치와 결과 사이의 모든 backdoor path를 막아야 한다.

그렇게 한다면, 남는 것은 T -> Y의 직접적인 효과 뿐일 것이다.

예시에서 '지능'수준을 통제할 수 있다면(동일 수준의 지능을 가지지만 교육 수준이 다른 사람들을 비교한다면) 모두 지능은 같기 때문에 결과의 차이는 오직 교육 수준의 차이에 의해 결정될 것이다.

교란 변수를 제거하기 위해서 우리는 처치와 결과의 모든 공통된 원인을 통제해야 한다.



```python
g = gr.Digraph()
g.edge("X", "T")
g.edge("X", "Y")
g.edge("T", "Y")

g.edge("Intelligence", "Educ"),
g.edge("Intelligence", "Wage"),
g.edge("Educ", "Wage")
g
```

하지만 우리는 모든 공통 원인을 항상 다 통제할 순 없다.

가끔 알려지지 않은 원인 혹은 알고 있지만 우리가 측정할 수 없는 원인도 있다.

'지능'의 경우는 후자에 해당한다.

***============================================================================***

***U*** 를 앞으로 측정되지 않은 변수를 가리킬 때 사용하겠다.

ex) 지능이 교육수준에 직접적인 영향은 미치지 못한다고 가정해보자.

지능은 SAT시험에서 얼마나 좋은 점수를 받을지 영향을 주지만, SAT는 좋은 대학에 갈 가능성을 열어서 교육 수준을 결정한다.

따라서, 우리가 측정 불가능한 '지능'을 통제할 수 없어도, 우리는 SAT점수는 통제할 수 있고 backdoor path를 막을 수 있다!

```python
g = gr.Digraph()
g.edge("X1", "T")
g.edge("T", "Y")
g.edge("X2", "T")
g.edge("X1", "Y")
g.edge("U", "X2")
g.edge("U", "Y")

g.edge("Family Income", "Educ")
g.edge("Educ", "Wage")
g.edge("SAT", "Educ")
g.edge("Family Income", "Wage")
g.edge("Intelligence", "SAT")
g.edge("Intelligence", "Wage")
g
```

위 그래프에서 X!, X2 혹은 가족소득(Family Income), SAT 에 대해 조건화하는 것은 처치와 결과 사이의 모든 backdoor path를 막기에 충분하다.

다시 말해, $(Y_0, Y_1) \perp T|X1, X2$ 이다.

따라서 모든 공통의 원인을 다 측정하지 못한다고 하더라도, 측정 불가능한 변수들이 처치에 미치는 영향을 매개하는 측정 가능한 변수를 통제해서 여전히 조건부 독립을 얻을 수 있다.

$(Y_0, Y_1) \perp T|X1, U$ 도 가지고 있지만, 우리는 U를 관측할 수 없기 때문에 이에 대해서는 조건화할 수 없다.

하지만!
측정 불가능한 변수가 처치와 결과의 직접적인 원인이 된다면 어떨까?

다음 예시(아래 그래프)를 살펴보자.

```python
g = gr.Digraph()
g.edge("X", "U")
g.edge("U", "T")
g.edge("T", "Y")
g.edge("U", "Y")

g.edge("Intelligence", "IQ")
g.edge("Intelligence", "SAT")
g.edge("Father's Educ", "Intelligence")
g.edge("Mother's Educ", "Intelligence")

g.edge("Intelligence", "Educ")
g.edge("Educ", "Wage")
g.edge("Intelligence", "Wage")

g
```

위 그래프를 보면, 지능은 교육과 소득의 원인이 된다. 

따라서, 처치인 교육과 결과인 임금의 관계 사이에 교란이 있는 것.

이 경우, 우리는 교란변수가 측정할 수 없기 때문에 통제할 수 없다. 그러나 우리는 교란변수의 프록시(proxy)가 되는 다른 측정 변수를 가지고 있다.

그 변수들은 backdoor path에 있지 않지만, 그것들을 통제하는 것이 편향을 줄이는데 도움이 된다(하지만 아예 제거할 수는 없다). 

이러한 변수들을 때로 ***대리 교란 변수*** 라고 칭한다.

위 그래프에서, 지능을 측정할 수 없지만, 지능의 몇 가지 원인이 되는 부모님의 교육 수준이나 지능이 효과를 미치는 SAT, IQ 점수 같은 것을 측정할 수 있다.

대리 변수들을 통제하는 것은 편향을 아예 제거하는 데 충분하지 않지만 도움이 된다.

# 선택 편향(Selection Bias)

교란 편향(confounding bias)를 제거하기 위해 측정할 수 있는 모든 것을 모델에 추가하는게 좋은 아이디어라고 생각할 수도 있다. 

하지만 다시 생각해 보자!

![image.png](./data/img/causal-graph/selection_bias.png)

교란 편향이 우리가 공통의 원인을 통제하지 않을 때 발생한다면, ***선택 편향*** 은 그것의 효과와 더 관련이 있다.

***선택 편향*** 은 우리가 통제해야하는 것보다 더 많은 변수를 통제할 때 발생한다.

처치와 잠재적 결과가 주변적 독립(marginally independent)이지만 collider에 대해 조건화하면 의존이 되는 것도 하나의 케이스이다.

아래 그래프 예시를 확인해보자

```python
g = gr.Digraph()
g.edge("T", "X")
g.edge("T", "Y")
g.edge("Y", "X")
g.node("X", "X", color="red")

g.edge("Educ", "Investments")
g.edge("Educ", "Wage")
g.edge("Wage", "Investments")

g
```

우리가 교육이 임금에 미치는 영향을 측정할 수 있게 무작위화가 가능해졌다고 상상해보자!

교란이 없음을 확신하기 위해 여러 변수를 통제한다.

그 중에서도 투자(Investments)도 통제한다. 하지만 투자는 교육과 임금의 공통 원인이 아니다. 대신 교육과 임금의 결과에 가깝다. 

또한, 많이 버는 사람이 투자도 더 많이 할 것이다.  투자가 collider이기 때문에, 이에 대해 조건화함으로써, 처치와 결과 사이에 두번째 경로를 열게 되고, 

이에 따라 직접적인 결과를 측정하기는 더 어려워진다.

이에 대해 생각하는 한 가지 방법은 투자를 통제함으로써 투자가 동일한 인구의 작은 그룹을 살펴본 ㄷ음 해당 그룹에 대한 교육의 효과를 찾는 것이다.

그러나 그렇게 함으로써 간접적 혹은 부주의로 인해 임금이 많이 변하지 않도록 만들고 있다.

결과적으로, 임금이 원래 상태처럼 변화하는 것을 허용하지 않기 때문에 교육이 임금을 어떻게 변화시키는지 알 수 없게 된다.



***==================================================================================================***

왜 이런 일이 발생하는지 이해하기 위해 투자와 교육이 오직 2가지 값만 가진다고 상상해보자.

    사람들이 투자를 하는지, 안하는지 and 교육을 받았는지 안받았는지.
    
투자를 한 사람들은 다음과 같이 나타낼 수 있다.  $E[Y_0|T=0, I=1] > E[Y_0|T=1, I=1]$
     
즉, 투자하는 사람들 중 교육 없이도 투자하는 사람들이 높은 수입을 얻기 위해 교육에 대해 더 독립적이다. 

이러한 이유로, 그 사람들이 받는 임금인 ${Wage}_{0}|T=0$은 교육받은 그룹이 교육받지 않았더라면 받았을 임금 ${Wage}_{0}|T=1$보다 더 높을 것이다.

비슷한 추론을 투자를 하지 않은 사람들에 대해서도 적용할 수 있는데, $ E[Y_0|T=0, I=0] > E[Y_0|T=1,I=0]$와 같다. 

교육을 받고도 투자하지 않은 사람들은 그들이 교육받지 않았더라면 교육받지 않고 투자하지 않은 사람들보다 더 낮은 임금을 받았을 가능성이 높다.

순수하게 그래프적 주장을 사용하면, 누군가 투자를 한다면 그가 높은 학력을 가진 것을 아는 것이 두 번째 원인인 임금을 잘 설명할 수 있다. 

투자에 대해 조건화했을 때, 교육 수준이 높은 것은 낮은 임금과 연관되어 있고, 우리는 음수 편향 $E[Y_0|T=0, I=i] > E[Y_0|T=1, I=i]$를 갖게 된다.

부가적인 설명으로써, 지금까지 우리가 논의한 모든 것들은 우리가 공통의 효과의 아무 후손(descendent)에 대해 조건화할 경우 모두 참이 된다.  

```python
g = gr.Digraph()
g.edge("T", "X")
g.edge("T", "Y")
g.edge("Y", "X")
g.edge("X", "S")
g.node("S", "S", color="red")
g
```

비슷한 일이 우리가 처치에 대한 매개변수(mediator)를 조건화할때에도 발생

매개변수란, 처치와 결과 사이에 존재하는 변수이다. 이 변수는 인과적 효과를 매개한다.

예를 들어, 교육 수준을 무작위화 할 수 있다고 가정해보자.

이 때, 어떤 사람이 화이트칼라 직업인지 아닌지만을 통제하기로 했다면 이러한 조건화는 인과효과추정에 편향을 만든다.

이번에는 collider에 대해 front door path를 열어서가 아니라, 처치자 작용하는 채널 중 하나를 막아버리기 때문이다.

아래 그래프에서, 화이트칼라 직업을 가지는 것은 교육이 더 높은 임금으로 가는 길에 위치해있다.

이를 통제함으로써, 우리는 이 채널을 막고 교육이 임금에 미치는 직접적인 효과만을 열린 채로 둘 수 있다.

```python
g = gr.Digraph()
g.edge("T", "X")
g.edge("T", "Y")
g.edge("X", "Y")
g.node("X", "X", color="red")

g.edge("educ", "white collar")
g.edge("educ", "wage")
g.edge("white collar", "wage")

g
```

잠재적 결과 주장을 하자면, 우리는 무작위화 때문에 편향이 0이 된다는 것을 알고 있다. 

$E[Y_0|T=0] - E[Y_0|T=1] = 0$. 

그러나, 만약 우리가 화이트칼라 직업을 가진 개인에 대해 조건화한다면, $E[Y_0|T=0, WC=1] > E[Y_0|T=1, WC=1]$ 이렇게 된다. 

이것은 교육 없이도 화이트칼라 직업을 얻을 수 있는 사람은 똑같은 직업을 얻기 위해 교육의 도움을 필요로하는 사람보다 더 열심히 일하는 사람일 가능성이 높기 때문이다. 

같은 방식으로 추론하면, $E[Y_0|T=0, WC=0] > E[Y_0|T=1, WC=0]$ 인데 이는 교육을 받고도 화이트칼라 직업을 얻지 못한 사람은 교육을 받지 않고 화이트칼라 직업을 얻지 못한 사람보다 덜 열심히 일하는 사람일 가능성이 높기 때문.

우리의 경우, 매개변수에 대해 조건화하는 것은 음수 편향을 유발한다. 

이는 교육의 효과를 실제보다 더 작아 보이게 만든다. 왜냐하면 인과 효과가 양수이기 때문이다. 만약 효과가 음인 경우 매개변수에 대한 조건화는 양의 편향을 가질 것. 모든 경우에 이러한 종류의 조건화는 효과를 실제보다 약해 보이게 만든다.

좀 더 단조롭게 표현해서 회사에서 두 명의 후보자 중 한 명을 선택해야 한다고 가정해 보자. 

둘 다 똑같이 인상적인 직업적 성취를 가지고 있지만, 하나는 더 높은 교육 학위를 가지고 있지 않다. 

어느 쪽을 선택해야 할까? 당연히 고등 교육을 받지 못한 사람을 선택해야 한다. 

왜냐하면 그는 다른 사람과 같은 일을 달성했지만 그에게 불리한 확률이 있었기 때문이다.
